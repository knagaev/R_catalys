```python
# проверка на уникальность 'Basic publication number main'
df_ps_bpnm_more_than_one = df_ps_prepared.groupby('Basic publication number main').filter(lambda x: len(x) > 1)

if len(df_ps_bpnm_more_than_one) > 0:
    raise ValueError("Дубликаты в 'Basic publication number main'")

df_ps_prepared.set_index(['Basic publication number main'], inplace=True)
    
df_ps_prepared.head() 
```

```python
#ищем записи, где в семействе нет базовой публикации (и добавляем её в семейство)
family_members = df_po_families['Family member']

df_earliest_not_in_family = df_po_families.loc[~df_po_families['Earliest publication number'].isin(family_members), ['Earliest publication number', 'Earliest publication number']]
df_earliest_not_in_family.columns = ['Earliest publication number', 'Family member']

#df_earliest_not_in_family.head()

df_po_families = pd.concat([df_po_families, df_earliest_not_in_family])

df_po_families.set_index(['Family member'], inplace=True)

df_po_families
```

```python
df_ps_po = df_ps_prepared.merge(df_po_families, left_index=True, right_index=True).merge(df_po, left_on='Earliest publication number', right_index=True, suffixes=('_ps', '_po'))
```

```python
po_ps_columns = ['Basic publication number', 'Family members_ps', 'Family members_po', 'Earliest publication number', 'Technology Relevance™', 'Market Coverage™', 'Competitive Impact™', 'Patent Asset Index™', 'Total number of citations received', 'Soundness', 'Novelty', 'Breadth', 'Focus of leading companies', 'Focus of leading countries']
df_source = df_ps_po.loc[:, po_ps_columns].sort_values('Basic publication number')
df_source.head()
```

```python
X_source = df_source[['Soundness', 'Novelty', 'Breadth', 'Focus of leading companies', 'Focus of leading countries']].values
y_source = df_source['Competitive Impact™'].values
```

```python
from sklearn.preprocessing import StandardScaler

sc_x = StandardScaler()
sc_y = StandardScaler()

X_std = sc_x.fit_transform(X_source)
y_std = sc_y.fit_transform(y_source.reshape(-1, 1)).flatten()
```

```python
from sklearn.preprocessing import QuantileTransformer, quantile_transform

y_trans = quantile_transform(y_source.reshape(-1, 1),
                             n_quantiles=100,
                             output_distribution='normal',
                             copy=True).squeeze()
```

```python
f, (ax0, ax1, ax2) = plt.subplots(1, 3)

ax0.hist(y_source, bins=100, **density_param)
ax0.set_ylabel('Probability')
ax0.set_xlabel('Target')
ax0.set_title('Source target distribution')

ax1.hist(y_std, bins=100, **density_param)
ax1.set_ylabel('Probability')
ax1.set_xlabel('Target')
ax1.set_title('Standardscaled target distribution')

ax2.hist(y_trans, bins=100, **density_param)
ax2.set_ylabel('Probability')
ax2.set_xlabel('Target')
ax2.set_title('Quantile_transformed target distribution')

f.suptitle("Calibration data: Competitive Impact", y=0.035)
f.tight_layout(rect=[0.05, 0.05, 0.95, 0.95])
```

```python
X = X_std
#y = y_std
y = y_trans
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
slr = LinearRegression()

slr.fit(X_train, y_train)

y_train_pred = slr.predict(X_train)
y_test_pred = slr.predict(X_test)
```

```python
slr.coef_
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, median_absolute_error, r2_score

print('MSE train: {:.3f}, test: {:.3f}'.format(
        mean_squared_error(y_train, y_train_pred),
        mean_squared_error(y_test, y_test_pred)))
print('R^2 train: {:.3f}, test: {:.3f}'.format(
        r2_score(y_train, y_train_pred),
        r2_score(y_test, y_test_pred)))
```

```python
from sklearn.linear_model import RidgeCV
regr = RidgeCV()
regr.fit(X_train, y_train)
y_train_pred = regr.predict(X_train)
y_test_pred = regr.predict(X_test)

print('MSE train: {:.3f}, test: {:.3f}'.format(
        mean_squared_error(y_train, y_train_pred),
        mean_squared_error(y_test, y_test_pred)))
print('R^2 train: {:.3f}, test: {:.3f}'.format(
        r2_score(y_train, y_train_pred),
        r2_score(y_test, y_test_pred)))
```
