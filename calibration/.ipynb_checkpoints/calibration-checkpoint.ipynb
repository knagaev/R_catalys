```python
import pandas as pd
```

```python
source_file_name = r'C:\PLR\catalys\calibration\ГПН_Ароформинг_Сила_PS_PO_ver_0_2.xlsx'
```

```python
with pd.ExcelFile(source_file_name) as xls_source_file:
  df_ps_strength = pd.read_excel(xls_source_file, 'PS_Strength')
  df_po_strength = pd.read_excel(xls_source_file, 'PO_Strength')

df_ps_strength.dropna(subset=['Basic publication number'], inplace=True)
df_po_strength.dropna(subset=['Earliest publication number'], inplace=True)

df_ps_strength.set_index('Basic publication number', inplace=True)
df_po_strength.set_index('Earliest publication number', inplace=True)
```

```python
# записи df_ps_strength, совпадающие по индексу с df_po_strength, копируются в df_ps_strength_result
df_ps_strength_result = df_ps_strength.loc[df_ps_strength.index.isin(df_po_strength.index)]
```

```python
# записи df_po_strength, совпадающие по индексу с df_ps_strength, копируются в df_po_strength_result
df_po_strength_result = df_po_strength.loc[df_po_strength.index.isin(df_ps_strength.index)]
```

```python
# семейства в df_ps_strength
df_ps_families = df_ps_strength['Family members'].str.split(';', expand=True).stack() \
    .str.strip().reset_index(drop=False).reset_index()#.set_index(0)
```

```python
# семейства в df_po_strength
df_po_families = df_po_strength['Family members'].str.split('\r\n', expand=True).stack() \
    .str.strip().reset_index(level=0, drop=False).reset_index().set_index(0)
```
